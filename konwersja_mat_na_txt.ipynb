```python
import pandas as pd
from scipy.io import loadmat
import os
```

```python
filename = 'USCB_catalog.mat' 
data = loadmat(f"data/mat_files/{filename}")

catalog_array = data['Catalog'][0] 

df_dict = {}

for entry in catalog_array:
    field_name = entry['field'][0]
    values = entry['val']
    
    if values.size > 0:
        df_dict[field_name] = values.flatten()

df = pd.DataFrame(df_dict)

print(df.shape)
display(df[:20])
```

```python
columns_for_matlab = ['Time', 'Lat', 'Long', 'Depth', 'ML'] 

kat_to_export = df[columns_for_matlab].copy()

clean_name = os.path.splitext(filename)[0]
kat_to_export.to_csv(f"data/txt_files/{clean_name}.txt", sep=' ', header=False, index=False)

print("Plik został wygenerowany pomyślnie")
```
