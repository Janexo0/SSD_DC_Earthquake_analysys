```python
import pandas as pd

filename = "USCB_Ziemowit_catalog"
file_path = f"data/txt_files/{filename}.txt"

df = pd.read_csv(file_path, sep=r'\s+', names=['time', 'lat', 'long', 'depth', 'mag'])

df.head()
```

```python
magnitude_threshold = 0.5

df_filtered = df[df['mag'] >= magnitude_threshold]

print(f"Rows before: {len(df)} | Rows after: {len(df_filtered)}")
df_filtered.head()
```

```python
clean_filename = "USCB_Ziemowit"
new_filename = f"{clean_filename}_MTR-{magnitude_threshold}.txt"
output_path = f"data/Ziemowit_txt/{new_filename}"

df_filtered.to_csv(output_path, sep=' ', header=False, index=False)

print(f"Filtered file successfully saved as: {output_path}")
```
